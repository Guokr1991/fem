#PML Test (Gaussian Excitation)

##Introduction

Perfect Matching Layers are used in a lot of numerical models (in house, Gianmarco's KZK code utilizes them), and LS-DYNA implemented them a few years ago for different material types.  Caterina's students started using them a few months ago, and I have implemented them for our models and wanted to see how they performed in models where the mesh might be a bit too small for a propagatin shear wave field.  I hope to show the propagating shear wave fields from a Gaussian excitation that encouters the top/bottom boundaries and eventually the edges to compare the old default non-reflecting boundary conditions with the PMLs.

##Methods

All of these models can be found in ```test/gauss[_pml]```, using the commands in ```run.sh```.  

````python
# import a bunch of useful modules
import scipy.io as sio

from matplotlib import pyplot as plt
%matplotlib inline

from IPython.display import display, Image
````

Here is a picture of the PML mesh, which by default utilizes layers of elements that are 5 elements thick on the top, bottom and outer faces (and can accomodate different symmetry models) for the PML (blue).  PMLs are now activated with the ```--pml``` argument when running ```bc.py``` and the non-reflecting BC has to be explicitely activated using ```--nonreflect```.

````python
display(Image(filename='./pml_mesh.png'))
````

````python
# load in the data
nonreflect = sio.loadmat('../gauss/res_sim.mat')
pml = sio.loadmat('./res_sim.mat')

def make_images(nonreflect, pml, ts):
    """
    generate displacement images of each BC at a specified time step, along with the difference
    """
    fig = plt.figure(figsize=(12,12))
    ax = fig.add_subplot(131)
    nr = ax.imshow(nonreflect['arfidata'][:,:,ts], \
                   extent = [nonreflect['lat'].min(), nonreflect['lat'].max(), nonreflect['axial'].max(), nonreflect['axial'].min()])
    ax.set_ylabel('Depth (mm)', fontsize=14, fontweight='bold')
    ax.set_title('Non-reflecting BC', fontsize=14, fontweight='bold')
    cbar = fig.colorbar(nr, shrink=0.5)
    
    ax = fig.add_subplot(132)
    p = ax.imshow(pml['arfidata'][:,:,ts], \
                  extent = [pml['lat'].min(), pml['lat'].max(), pml['axial'].max(), pml['axial'].min()])
    ax.set_xlabel('Lateral (mm)', fontsize=14, fontweight='bold')
    ax.set_title('PML BC', fontsize=14, fontweight='bold')
    cbar = fig.colorbar(p, shrink=0.5, pad=0.1)

    pml_nr_diff = pml['arfidata'][:,:,ts] - nonreflect['arfidata'][:,:,ts]
    ax = fig.add_subplot(133)
    d = ax.imshow(pml_nr_diff, \
                  extent = [nonreflect['lat'].min(), nonreflect['lat'].max(), nonreflect['axial'].max(), nonreflect['axial'].min()], \
                  vmin=-5, vmax=5,
                  cmap='seismic')
    ax.set_title('Difference', fontsize=14, fontweight='bold')
    cbar = fig.colorbar(d, ax=ax, shrink=0.5)
    cbar.set_label('Displacement (um)', fontsize=14, fontweight='bold')
````

##Results

###Time Step 5

````python
make_images(nonreflect, pml, 5)
````

###Time Step 10

````python
make_images(nonreflect, pml, 10)
````

###Time Step 15

````python
make_images(nonreflect, pml, 15)
````

###Time Step 19

````python
make_images(nonreflect, pml, 19)
````

###Run Time

The PML model took ~2.5x longer to run than the non-reflecting BC model.  This was not thoroughly tested, but is in the roughly-expected ballpark.

##Discussion

This mesh should jump out as being too small axially for the size of the source, and too small in the lateral and elevation dimensions for the stiffness and time of the simulation, but this was done to amplify potential differences between the different BCs.

When the mesh is large enough in the lateral and elevation dimensions such that the shear wave isn't interacting with the outer faces, then there are only differences at the top and bottom face boundaries, which in this case were relatively close to the excitation source.  By time step 10, you can definitely see differences in the upper and lower "tails" for the shear wave front, and there are shear wave amplitude differences.  *This is important if attenuation is being estimated from amplitude decay.*  While group shear wave speed doesn't appear to be impacted, that hasn't been quantified yet.  These effects could also lead to difference in VE materials.

All bets are off by time step 19, but you can see some of the artifacts in the non-reflecting BC case that we see in our "good" models that run for too long.

##Conclusions

* No qualitative difference in group shear wave speed.
* Need to evluate impact on shear wave frequency content.
* Need to evluate impact on VE materials, which is trickier to do since a native VE PML doesn't exist, but a close elastic approximation can be made.
* Run times are longer with the PMLs, but meshes can be made smaller, especially in the elevation dimension without worry about reflections coming back into the imaging plane.
