#Impact of PMLs in VE Liver Models

##Introduction
--------------

Evaluating the impact of PMLs on the VE liver models that we ran for the IEEE manuscript that proposed methods for evaluating viscous effects in our *in vivo* liver datasets.

##Methods
---------

The 2 "extreme" material parameter sets from the VE liver were analyzed:

###Healthy Liver
G0 = 10 kPa
Ginfinity = 2 kPa
beta = 6666.67 /s

These give, approximately, c(200 Hz) = 1.9 m/s and dc/df = 3.0 m/s/kHz.

###Fibrotic Liver
G0 = 20 kPa
Ginfinity = 4 kPa
beta = 4000 /s

These give, approximately, c(200 Hz) = 2.9 m/s and dc/df = 5.5 m/s/kHz.

Velocity fields at time points are calculated, along with velocity difference fields (%).

```python
# import a bunch of useful modules
import scipy.io as sio

from matplotlib import pyplot as plt
%matplotlib inline

from IPython.display import display, Image
```

```python
def make_images(nonreflect, pml, ts):
    """
    generate displacement images of each BC at a specified time step, along with the difference
    """
    import numpy as np
    
    fig = plt.figure(figsize=(12,12))
    fig.suptitle('Time Step %i' % ts, fontsize=18, fontweight='bold')
    ax = fig.add_subplot(131)
    NRvel = np.diff(nonreflect['arfidata'])*1e3
    nr = ax.imshow(NRvel[:,:,ts], \
                   extent = [nonreflect['lat'].min(), nonreflect['lat'].max(), nonreflect['axial'].max(), nonreflect['axial'].min()])
    ax.set_ylabel('Depth (mm)', fontsize=14, fontweight='bold')
    ax.set_title('Non-reflecting BC', fontsize=14, fontweight='bold')
    cbar = fig.colorbar(nr, shrink=0.6)
    
    ax = fig.add_subplot(132)
    PMLvel = np.diff(pml['arfidata'])*1e3
    p = ax.imshow(PMLvel[:,:,ts], \
                  extent = [pml['lat'].min(), pml['lat'].max(), pml['axial'].max(), pml['axial'].min()])
    ax.set_xlabel('Lateral (mm)', fontsize=14, fontweight='bold')
    ax.set_title('PML BC', fontsize=14, fontweight='bold')
    cbar = fig.colorbar(p, shrink=0.6)

    pml_nr_diff = 100*(PMLvel[:,:,ts] - NRvel[:,:,ts])/np.amax(PMLvel[:,:,ts])
    ax = fig.add_subplot(133)
    d = ax.imshow(pml_nr_diff, \
                  extent = [nonreflect['lat'].min(), nonreflect['lat'].max(), nonreflect['axial'].max(), nonreflect['axial'].min()], \
                  vmin=-30, vmax=30,
                  cmap='seismic')
    ax.set_title('Difference', fontsize=14, fontweight='bold')
    cbar = fig.colorbar(d, ax=ax, shrink=0.6)
    cbar.set_label('Velocity (um/ms)', fontsize=14, fontweight='bold')
    fig.tight_layout()
```

##Results
---------

###Healthy Liver
----------------

```python
nonreflect = sio.loadmat('/radforce/fem/ve_liver/G010.0kPa/GI2.0kPa/BETA666.7/res_sim.mat')
pml = sio.loadmat('/radforce/fem/ve_liver/G010.0kPa/GI2.0kPa/BETA666.7_PML/res_sim.mat')
for ts in range(0,148,25):
    make_images(nonreflect, pml, ts)
```

###Fibrotic Liver

```python
nonreflect = sio.loadmat('/radforce/fem/ve_liver/G020.0kPa/GI4.0kPa/BETA4000.0/res_sim.mat')
pml = sio.loadmat('/radforce/fem/ve_liver/G020.0kPa/GI4.0kPa/BETA4000.0_PML/res_sim.mat')
for ts in range(0,148,25):
    make_images(nonreflect, pml, ts)
```

##Discussion
-------------

Definitely more relative error at each time step than I was expecting (hoping for), but most of the largest regions of error are away from the "focal depth" of the Gaussian excitation.  That being said, for the purposes of higher-order analysis, I'm almost certain that we'll see spectral differences... just not sure how significant they will be.

It should be noted for the fibrotic liver case that we didn't use those later time steps in the analysis b/c the shear wave hits the edge.

##Conclusions
-------------
* Definite differences with the PMLs, with early difference being at the top/bottom corners of the models, but also changes in the depth-of-field.
* Need to quantitatively evaluate:
 * Group shear wave speeds
 * Shear wave spectral content
 * Phase velocity slope
 * Phase velocity @ 200 Hz
